# Seaborn
Seaborn is a Python data visualization library built on top of Matplotlib. It makes it easier and more attractive to create statistical graphics.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print(sns.get_dataset_names())
df = sns.load_dataset("tips")
df
```

```python
sns.displot(df["total_bill"],kde=False) #KDE flag is set to false as a result the representation of the kernal estimation plot will be removed and only histogram is plotted
plt.show                                # KDE- Kernal Density Estimation is the way to estimate the probability density function of continue random variable. It is used for non matric analyses.
```

```python
#jointplot() is a Seaborn function that helps you visualize the relationship between two numeric variables along with their individual distributions.
#ðŸ“Œ Use Cases: Correlation analysis. Outlier detection. Visualizing clusters by category (hue). Including in EDA reports
import seaborn as sns
sns.jointplot(x="total_bill",y="tip",data=df) 
plt.show
```

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind="hex") 
plt.show()
```

```python
# barplot() is a Seaborn function used to draw bar charts that show the average value of a numeric variable for each category in a categorical variable.
# ðŸ“Œ Use Cases: Comparing category-wise averages (e.g., sales by region). Displaying group-wise summaries. Visualizing statistical estimates
sns.barplot(x="total_bill",y="tip",data=df) 
plt.show
```

```python
#countplot() is a Seaborn function used to show the count (frequency) of observations for each category of a categorical variable.
# ðŸ“Œ Use Cases: Counting how many customers chose each payment method.Showing gender distribution in a dataset.Checking data balance (e.g., spam vs non-spam emails).
sns.countplot(x="total_bill",data=df) 
plt.show
```

```python
# pointplot() shows the mean value of a numerical variable for each category along with confidence intervals (error bars). It's great for comparing group means.
#ðŸ“Œ Use Cases: Comparing average sales by category,Visualizing trends in scores across groups,Showing effect of treatments in experiments.
sns.pointplot(x="total_bill",y="tip",data=df) 
plt.show
```

```python
# lineplot() is used to draw line graphs representing trends or changes over continuous data (like time, quantity, etc.). It shows data points connected by lines and is perfect for time series and trend analysis.
# ðŸ“Œ Use Cases: Time-series analysis (sales over months, temperature over days).Trend detection (growth patterns, decline patterns),Comparing categories over time.
sns.lineplot(x="total_bill",y="tip",data=df) 
plt.show
```

```python
# palplot() is a Seaborn function used to visually display a color palette. Itâ€™s helpful when you want to see the actual colors in a palette you plan to use in plots.
#ðŸ§  Use Case: Helps preview and select colors before applying them to plots. Useful when creating custom themes or designing consistent visuals.
sns.palplot(sns.color_palette()) #plots the color palette as a horizontal array
plt.show()
```

```python
print(sns.get_dataset_names())
df = sns.load_dataset("penguins")
df
```

```python
#The head() function is used to view the first few rows of a DataFrame â€” super useful when you want a quick look at your data.
df.head()
```

```python
#The describe() function in Pandas is used to generate summary statistics of your data â€” numerical or categorical â€” quickly.
df.describe()
```

```python
#The info() function gives a summary of your DataFrame, including data types, non-null counts, and memory usage.
df.info()
```

```python
# The histplot() function in Seaborn is used to plot histograms â€” it shows the distribution of a single numerical variable by dividing the data into bins
sns.histplot(df["body_mass_g"])
plt.show()
```

```python
# The kdeplot() function is used to plot the Kernel Density Estimate (KDE) â€” a smooth, continuous curve that shows the probability distribution of a numeric variable.
sns.kdeplot(df["body_mass_g"])
plt.show()
```

```python
# The scatterplot() function is used to create a scatter plot, which shows the relationship between two numeric variables using dots.
sns.scatterplot(x="flipper_length_mm",y="body_mass_g",data=df)
plt.show()
```

```python
# The lineplot() function is used to draw line graphs, especially useful for showing trends or changes over time.
sns.lineplot(df["flipper_length_mm"])
plt.show()
```

```python
# The pairplot() function helps you visually explore relationships between multiple numerical variables in your dataset.
sns.pairplot(df)
plt.show()
```

```python
# The boxplot() function is used to visualize the distribution, spread, and outliers of a numerical feature using a box-and-whisker plot Whiskers (typically 1.5Ã—IQR from Q1 & Q3).
sns.boxplot(x="species",y="body_mass_g",data=df)
plt.show()
```

```python
# The countplot() function is used to show the count (frequency) of categorical values in a column. Itâ€™s a quick way to visualize how many times each category appears.
sns.countplot(df["species"])
plt.show()
```

```python
# The violinplot() is used to visualize the distribution of numerical data across different categories. It combines a boxplot and a KDE (kernel density estimate) plot into a single visualization
sns.violinplot(x="sex",y="species",data=df)
plt.show()
```

```python
# The heatmap() function is used to visualize data in matrix form with colors representing values. Itâ€™s perfect for showing correlations, confusion matrices, or any 2D data.
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.show()
```

```python
# The palplot() function in Seaborn is used to display a color palette as a horizontal strip of colors. Itâ€™s helpful when you want to visualize the colors from a palette before using them in your plots.
sns.palplot(sns.color_palette())
plt.show()
```

```python
# The set_palette() function in Seaborn is used to set the default color palette for all plots. It changes the look and feel of your plots by applying a specific color scheme.
sns.set_palette("pastel")
plt.show()
```

```python
# Use a Custom Color Palette
# You can create a custom color palette using sns.color_palette().
custom_colors = ["#FF5733", "#33FF57", "#3357FF"]  # Define custom colors
sns.set_palette(custom_colors)
# Example: Boxplot with custom colors
sns.boxplot(x="species", y="body_mass_g", data=df)
plt.title("Body Mass Distribution with Custom Palette")
plt.show()
```

```python
# 2. Adjust Figure Size
# You can adjust figure size using plt.figure(figsize=(width, height)).
plt.figure(figsize=(10, 6))  # Width=10 inches, Height=6 inches
sns.histplot(df["body_mass_g"], kde=True) #The histplot() function in Seaborn is used to draw a histogram, which shows the distribution of a single numerical variable by dividing it into intervals (bins)
plt.title("Histogram with Custom Figure Size")
plt.show()
```

```python
# Adjusting Plot Styles
# Seaborn provides different styles to enhance visual appeal.
# Available styles: "white", "dark", "whitegrid", "darkgrid", "ticks"
sns.set_style("darkgrid")
#The scatterplot() function in Seaborn is used to draw a scatter plot, which shows the relationship between two numeric variables.
sns.scatterplot(x="flipper_length_mm", y="body_mass_g", hue="species", data=df)
plt.title("Scatter Plot with Darkgrid Style")
plt.show()

```

```python
#Remove Spines for a Cleaner Look
sns.set_style("white")  # Removes grid
sns.boxplot(x="species", y="body_mass_g", data=df)
# The sns.despine() function is used to remove the borders (spines) from the plots to make them cleaner and more professional-looking.
sns.despine(left=True, bottom=True)  # Removes spines
plt.title("Boxplot with Removed Spines")
plt.show()
```

```python
# Changing Themes with sns.set()
# You can set global styles, font sizes, and colors using sns.set().
sns.set(
    style="whitegrid",
    context="talk",  # "paper", "notebook", "talk", "poster"
    palette="coolwarm",
)
# The violinplot() is used to visualize the distribution of numerical data across different categories. It combines a boxplot and a KDE (kernel density estimate) plot into a single visualization.
sns.violinplot(x="species", y="body_mass_g", data=df)
plt.title("Violin Plot with Custom Global Style")
plt.show()
```

```python
#Customize Fonts and Titles
sns.set(font="Arial", font_scale=1.2)
  # Adjust font size

plt.figure(figsize=(8, 5))
sns.histplot(df["body_mass_g"], kde=True)
plt.title("Custom Font and Scale")
plt.show()
```

```python
#The sns.PairGrid() function is used to create a grid of subplots for plotting pairwise relationships between multiple variables in a DataFrame. It gives you more custom control compared to pairplot().
sns.PairGrid(df)
plt.show()
```
